**Data Visualization Module of Data Science**



*   Importing necessary libraries



```python
import matplotlib.pyplot as plt
import seaborn as sns
```



*   Data Visualization Class having some data visualization techniques




```python
class DataVisualization:
    def __init__(self, df):
        """
        Initializes the DataVisualization object with a pandas DataFrame.

        Parameters:
        df (pandas DataFrame): The DataFrame to be visualized.
        """
        self.df = df

    def plot_histogram(self, column):
        """
        Plots a histogram of the specified column in the DataFrame.

        Parameters:
        column (str): The name of the column to be plotted.
        """
        sns.histplot(data=self.df, x=column)
        plt.show()

    def plot_boxplot(self, column):
        """
        Plots a boxplot of the specified column in the DataFrame.

        Parameters:
        column (str): The name of the column to be plotted.
        """
        sns.boxplot(data=self.df, y=column)
        plt.show()

    def plot_scatter(self, x_column, y_column):
        """
        Plots a scatter plot of the specified x and y columns in the DataFrame.

        Parameters:
        x_column (str): The name of the x column to be plotted.
        y_column (str): The name of the y column to be plotted.
        """
        sns.scatterplot(data=self.df, x=x_column, y=y_column)
        plt.show()

    def plot_heatmap(self):
        """
        Plots a heatmap of the correlations between columns in the DataFrame.
        """
        sns.heatmap(self.df.corr(), annot=True, cmap='coolwarm')
        plt.show()

    def plot_pairplot(self):
        """
        Plots a pairplot of the columns in the DataFrame.
        """
        sns.pairplot(data=self.df)
        plt.show()

    def plot_barplot(self, x_column, y_column):
        """
        Plots a barplot of the specified x and y columns in the DataFrame.

        Parameters:
        x_column (str): The name of the x column to be plotted.
        y_column (str): The name of the y column to be plotted.
        """
        sns.barplot(data=self.df, x=x_column, y=y_column)
        plt.show()

    def plot_countplot(self, column):
        """
        Plots a countplot of the specified column in the DataFrame.

        Parameters:
        column (str): The name of the column to be plotted.
        """
        sns.countplot(data=self.df, x=column)
        plt.show()
```



*   Testing on sample data



```python
import pandas as pd

data=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

```python
# Create a DataVisualization object with the preprocessed DataFrame
visualization = DataVisualization(data)

# Plot a histogram of a column
visualization.plot_histogram("Hours")

```

```python
# Plot a scatter plot of two columns
visualization.plot_scatter("Hours", "Scores")

```

```python
# Plot a heatmap of correlations between columns
visualization.plot_heatmap()

```

```python
# Plot a pairplot of all columns
visualization.plot_pairplot()
```
